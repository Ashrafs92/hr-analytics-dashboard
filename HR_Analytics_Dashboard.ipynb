# HR Analytics Dashboard Project
This notebook explores attrition trends in HR data.

```
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("hr_analytics_data.csv")
df.head()
```

```
# Attrition by Department
df.groupby("Dept")["Attrition"].value_counts().unstack().plot(
    kind="bar", stacked=True, figsize=(8,5), title="Attrition by Department"
)
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.show()
```

```
# Attrition by Age Group
df['AgeGroup'] = pd.cut(df['Age'], bins=[20,30,40,50,60],
                        labels=["20-30","31-40","41-50","51-60"])

df.groupby("AgeGroup")["Attrition"].value_counts().unstack().plot(
    kind="bar", stacked=True, figsize=(8,5), title="Attrition by Age Group"
)
plt.xlabel("Age Group")
plt.ylabel("Employee Count")
plt.show()
```
